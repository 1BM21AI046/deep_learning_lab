```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models

# Generate some sample data
np.random.seed(0)
X = np.linspace(-5, 5, 100).reshape(-1, 1)
y = np.sin(X)

# Define the architecture of the neural network
def create_model(input_shape, hidden_units, num_classes, activation_hidden='relu', activation_output='sigmoid'):
    model = models.Sequential()

    # Input layer
    model.add(layers.Flatten(input_shape=input_shape))

    # Hidden layers
    for units in hidden_units:
        model.add(layers.Dense(units, activation=activation_hidden))

    # Output layer
    model.add(layers.Dense(num_classes, activation=activation_output))

    return model

# Create the model
input_shape = X.shape[1:]
hidden_units = [64, 32]
num_classes = 1
model = create_model(input_shape, hidden_units, num_classes)

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X, y, epochs=100, verbose=0)

# Plot input vs. output
plt.figure(figsize=(8, 6))
plt.plot(X, y, label='True Function')
plt.plot(X, model.predict(X), label='Neural Network Prediction')
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Input vs. Output')
plt.legend()
plt.grid(True)
plt.show()
```
